```python
def create_webdriver(url):
    driver = webdriver.Edge()
    driver.get(url)
    driver.maximize_window()
    return driver

def crawl_url_product(driver):
    try:
        box_product = driver.find_element(By.CLASS_NAME, "listproduct").find_elements(By.TAG_NAME, "li")
        print(len(box_product))
    except:
        print("error")
    
    url_product = []
    for i,item in enumerate(box_product):
        try:
            url_item = item.find_element(By.TAG_NAME, "a").get_attribute("href")
            url_product.append(url_item)
        except:
            print(i)
            pass
    return box_product, url_product
```

```python
url_tablet = "https://www.thegioididong.com/may-tinh-bang#c=522&o=17&pi=1"
driver_tablet = create_webdriver(url_tablet)
```

```python
box_tablet_product, url_tablet_product = crawl_url_product(driver_tablet)
```

```python
with open("url_tablet.txt", "w") as f:
    for item in url_tablet_product:
        f.write(item+"\n")
```
